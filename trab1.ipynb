```python
ufTargets = "['AC', 'AL', 'PR', 'TO']" # this is for fitring only the study group
```

## i. Qual  dos 4  estados,  levando  em  consideração  apenas  a  1ª  dose  da  vacina, possui  o  maior índice de vacinação por habitante?

```python
question_i = dataFixed.query('vacina_descricao_dose == "1a dose"').groupby(by='paciente_endereco_uf').count().query("paciente_endereco_uf == "+ufTargets)  
question_i.reset_index(inplace = True)
question_i
```

```python
fig = px.bar(question_i,x='paciente_endereco_uf',y='document_id',labels={'paciente_endereco_uf':'UF','document_id':'N Vaccine'}, title='Number of vaccinated by state') 
fig.show()
```

## ii. Qual foi a idade que mais foi vacinada em cada Estado? Pode-se dizer que há inconsistências nos dados analisados?

```python
ufTargets = ['AC', 'AL', 'PR', 'TO']
```

```python
question_ii = dataFixed.groupby(by=['paciente_endereco_uf','paciente_idade']).count().sort_values('document_id',ascending=False)
question_ii
```

```python
ages = []
for uf in ufTargets:
    aux = question_ii.loc[uf].reset_index().head(1).to_numpy()[0][0:2]
    print('UF: {:10}\tIdade: {:10}\tNumero: {:10}'.format(uf,aux[0],aux[1]))
    ages.append([uf,aux[0],aux[1]])
```

## v.Pode-se afirmar que se vacinou mais homens ou mulheres?

```python
question_v = dataFixed.query('vacina_descricao_dose == "1a dose"').groupby(by=['paciente_endereco_uf','paciente_enumSexoBiologico']).count().sort_values('document_id',ascending=False).loc[ufTargets].reset_index().groupby(by='paciente_enumSexoBiologico').sum()  
question_v
```

```python
print('Mulheres vacinadas:{}\nHomens vacinados:{}'.format(question_v.document_id.to_numpy()[0],question_v.document_id.to_numpy()[1]))  
```
